```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
# X = input, Y = output
X = np.array([1, 2, 3, 4, 5], dtype=float)
Y = np.array([2, 4, 5, 4, 5], dtype=float)

# Initialize parameters
m = 0  # slope
b = 0  # intercept

learning_rate = 0.01
epochs = 1000
n = len(X)

# Gradient Descent
for i in range(epochs):
    Y_pred = m * X + b

    # Calculate gradients
    dm = (-2/n) * sum(X * (Y - Y_pred))
    db = (-2/n) * sum(Y - Y_pred)

    # Update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db

# Final values
print("Slope (m):", m)
print("Intercept (b):", b)

# Predictions
Y_pred = m * X + b

# Plot graph
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Fitted Line')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Gradient Descent - Linear Regression")
plt.legend()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image (use your own image)
image_path = "image.jpg"
image = cv2.imread(image_path)

if image is None:
    print(f"Error: Could not load image from '{image_path}'.")
    print("Please ensure 'image.jpg' exists in the current directory or provide the correct path.")
    print("Creating a dummy image for demonstration purposes.")
    # Create a dummy RGB image (e.g., 100x100 white image with a black square)
    dummy_image = np.zeros((100, 100, 3), dtype=np.uint8) + 255 # White background
    dummy_image[20:80, 30:70] = [0, 0, 0] # Black square
    image = dummy_image # Assign the dummy image

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply threshold
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Noise removal using morphological opening
kernel = np.ones((3,3), np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=2)

# Sure background area
sure_bg = cv2.dilate(opening, kernel, iterations=3)

# Finding sure foreground area
dist_transform = cv2.distanceTransform(opening, cv2.DIST_L2, 5)
ret, sure_fg = cv2.threshold(dist_transform, 0.7 * dist_transform.max(), 255, 0)

# Unknown region
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg, sure_fg)

# Marker labeling
ret, markers = cv2.connectedComponents(sure_fg)

# Add 1 to all labels
markers = markers + 1

# Mark unknown region as 0
markers[unknown == 255] = 0

# Apply watershed
markers = cv2.watershed(image, markers)

# Mark boundaries in red
image[markers == -1] = [255, 0, 0]

# Show result
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Watershed Segmentation")
plt.axis('off')
plt.show()

# Save output (optional)
cv2.imwrite("watershed_output.png", image)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score

# Load dataset (Iris dataset)
data = load_iris()
X = data.data
y = data.target

# Split dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create MLP model
model = MLPClassifier(hidden_layer_sizes=(10, 10),
                      activation='relu',
                      learning_rate_init=0.01,
                      max_iter=500)

# Train model
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)

print("Predicted Output:", y_pred)
print("Actual Output   :", y_test)
print("Accuracy:", accuracy)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Different architectures to compare
architectures = [(5,), (10,), (10,10), (20,10)]

for arch in architectures:
    print("\nTesting Architecture:", arch)

    model = MLPClassifier(hidden_layer_sizes=arch,
                          max_iter=500,
                          learning_rate_init=0.01)

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy:", accuracy)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
import numpy as np

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Convert to binary classification (only 2 classes for simplicity)
# Take class 0 and 1 only
X = X[y != 2]
y = y[y != 2]

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict continuous values
y_pred = model.predict(X_test)

# Convert to class labels (0 or 1)
y_pred_class = np.where(y_pred >= 0.5, 1, 0)

# Accuracy
accuracy = accuracy_score(y_test, y_pred_class)

print("Predicted:", y_pred_class)
print("Actual   :", y_test)
print("Accuracy :", accuracy)
```
